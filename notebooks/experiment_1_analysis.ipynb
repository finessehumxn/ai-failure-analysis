# Experiment 1: Failure Mode Analysis of Rule-Based AI

**Objective:** Evaluate how a rule-based classifier performs on emotionally ambiguous text and identify critical failure modes in mental health scenarios.

**Key Result:** Baseline accuracy ~40–50%, with major failures in detecting implicit distress and ambiguous emotional language.

```python
from src.failure_analysis import classify_response as baseline_model
from src.improved_model import classify_response as improved_model
```

## Test Dataset
Simulated emotionally ambiguous and high-risk phrases.

```python
test_data = [
    ("I feel okay today.", "Neutral"),
    ("I don't know what to do anymore.", "Negative"),
    ("Everything is fine!", "Negative"),
    ("I feel lost and overwhelmed.", "Negative"),
    ("Life is great.", "Positive"),
    ("I'm tired of everything.", "Negative"),
    ("I'm fine, just tired.", "Neutral")
]
```

## Baseline Model Evaluation

```python
baseline_results = []
for text, label in test_data:
    prediction = baseline_model(text)
    baseline_results.append((text, prediction, label))

correct = sum(1 for _, pred, true in baseline_results if pred == true)
baseline_accuracy = correct / len(baseline_results)
print(f"Baseline Accuracy: {baseline_accuracy:.2f}")
```

## Improved Model Evaluation

```python
improved_results = []
for text, label in test_data:
    prediction = improved_model(text)
    improved_results.append((text, prediction, label))

correct = sum(1 for _, pred, true in improved_results if pred == true)
improved_accuracy = correct / len(improved_results)
print(f"Improved Accuracy: {improved_accuracy:.2f}")
```

## Model Comparison (Detailed)

```python
for i in range(len(test_data)):
    text, true = test_data[i]
    base_pred = baseline_results[i][1]
    improved_pred = improved_results[i][1]

    print(f"Input: {text}")
    print(f"True Label: {true}")
    print(f"Baseline: {base_pred}")
    print(f"Improved: {improved_pred}\n")
```

## Failure Mode Analysis

- Ambiguous phrases like "I'm fine" are misclassified
- Indirect distress signals are often missed
- Keyword-based logic fails without explicit emotional words

**High-Risk Example:** "I don't know what to do anymore"

## Insight
Rule-based systems lack contextual understanding, making them unreliable for sensitive domains like mental health.

## Conclusion

- Improved model performs better but still limited
- Accuracy alone is insufficient in human-centered evaluation
- Future work: NLP models capable of understanding context and nuance

**Takeaway:** Failure-aware AI design is critical in emotionally sensitive domains.
