# Experiment 1: Baseline Model Analysis

This notebook explores the behavior of a simple rule-based classifier used to analyze emotionally ambiguous text.

The goal is to identify failure modes and evaluate how the model performs in human-centered scenarios, particularly in mental health contexts.

```python
from src.failure_analysis import classify_response as baseline_model
from src.improved_model import classify_response as improved_model
```

```python
test_data = [
    ("I feel okay today.", "Neutral"),
    ("I don’t know what to do anymore.", "Negative"),
    ("Everything is fine!", "Negative"),
    ("I feel lost and overwhelmed.", "Negative"),
    ("Life is great.", "Positive")
]
```

## Baseline Model Evaluation

```python
baseline_results = []

for text, label in test_data:
    prediction = baseline_model(text)
    baseline_results.append((text, prediction, label))

baseline_results
```

```python
correct = sum(1 for _, pred, true in baseline_results if pred == true)
accuracy = correct / len(baseline_results)
accuracy
```

## Model Comparison

```python
comparison_results = []

for text, label in test_data:
    base_pred = baseline_model(text)
    improved_pred = improved_model(text)

    comparison_results.append((text, base_pred, improved_pred, label))

for text, base, improved, true in comparison_results:
    print(f"Input: {text}")
    print(f"Baseline: {base} | Improved: {improved} | Actual: {true}\n")
```

## Observations

- The baseline model struggles with emotionally ambiguous language
- The improved model better captures implicit distress signals
- Introducing an "Unknown" category reduces overconfident misclassification

## Key Insight

Even with improvements, rule-based systems fail to capture deeper emotional context, reinforcing the need for more advanced NLP approaches.
